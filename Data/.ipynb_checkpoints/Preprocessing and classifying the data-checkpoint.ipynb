## 1. Preprocessing the data
### Fake is 1 Real is 0 for all subsequent datasets

```python
import pandas as pd
```

```python
buzzfeedfake = pd.read_csv('BuzzFeed_fake_news_content.csv')
buzzfeedreal = pd.read_csv('BuzzFeed_real_news_content.csv')
fakes = pd.read_csv('FA-KES-Dataset.csv', encoding='unicode_escape')
fake = pd.read_csv('Fake.csv')
real = pd.read_csv('True.csv')
politifact = pd.read_csv ('PolitiFact_real_news_content.csv')
```

```python
buzzfeedfake['fake'] = 1
buzzfeedfake = buzzfeedfake.drop(columns =  ['id', 'url', 'top_img', 'authors', 'source', 'publish_date', 'movies', 'images', 'canonical_link', 'meta_data'])
buzzfeedfake.head()
```

```python
buzzfeedreal['fake'] = 0
buzzfeedreal = buzzfeedreal.drop(columns = ['id', 'url', 'top_img', 'authors', 'source', 'publish_date', 'movies', 'images', 'canonical_link', 'meta_data'])
buzzfeedreal.head()
```

```python
fakes = fakes.drop(columns = ['unit_id', 'source', 'date', 'location'])
fakes = fakes.rename(columns = {'labels':'fake', 'article_title': 'title', 'article_content':'text'})
fakes['fake'] = 1
fakes.head()
```

```python
fake = fake.drop(columns = ['subject', 'date'])
fake['fake'] = 1
fake.head()
```

```python
real = real.drop(columns = ['subject', 'date'])
real['fake'] = 0
real.head()
```

```python
politifact = politifact.drop(columns = ['id', 'url', 'top_img', 'authors', 'source', 'publish_date', 'movies', 'images', 'canonical_link', 'meta_data'])
politifact['fake'] = 1
politifact.head()
```

```python
final = pd.concat((buzzfeedfake, buzzfeedreal, fake, fakes, real, politifact))
```

```python
final
```

### 2. Making the words all lowercase

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
import re
from nltk.stem.porter import PorterStemmer
from nltk.stem import SnowballStemmer
```

```python
def cleanData(text):
    text = re.sub("((\S+)?(http(s)?)(\S+))|((\S+)?(www)(\S+))|((\S+)?(\@)(\S+)?)", "", text) ## removes all punctuation
    text = re.sub("[^a-zA-Z0-9!?%()<>{}:;\"\',.\[\] ]", "", text)
    text = text.lower()
    text = nltk.word_tokenize(text)
    return text
```

```python
final['title'] = final['title'].apply(cleanData)
final['text'] = final['text'].apply(cleanData)
```

```python
final.head()
```

```python
stop_words = stopwords.words('english')
def removeStopWords(text):
    return [word for word in text if word not in stop_words]
```

```python
final['title'] = final['title'].apply(removeStopWords)
final['text'] = final['text'].apply(removeStopWords)
```

```python
final
```

```python
stemmer = PorterStemmer()
def stemming(text):
    try:
        text = [stemmer.stem(word) for word in text]
        text = [word for word in text if len(word) > 1] # make sure we have no 1 letter words
    except IndexError:
        pass
    return text
```

```python
snowball = SnowballStemmer(language='english')
def sbstemming(text):
    try:
        text = [snowball.stem(word) for word in text]
        text = [word for word in text if len(word) > 1] # make sure we have no 1 letter words
    except IndexError:
        pass
    return text
```

```python
final['title'] = final['title'].apply(sbstemming)
final['text'] = final['text'].apply(sbstemming)
```
